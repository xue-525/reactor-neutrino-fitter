```python
from iminuit import Minuit
from fitter.src.fitter.analysis.fitter import Fitter
import psutil
import os
import time
import torch
from fitter.src.fitter.config import GlobalConfig as gcfg

start_time = time.perf_counter()
process = psutil.Process(os.getpid())
initial_cpu = process.cpu_times()
gseed = 50  # 0 for Asimov >0 for toy
year = 50/365.25 # 1年统计量
# gcfg.use_poisson = True  # 是否使用泊松分布

junosyst = Fitter(year,        
        n_E_nu_bins=5600,  # 默认对应0.002 MeV步长 (13.0-1.8)/5600≈0.002
        n_E_dep_bins=5600,  # 默认对应0.002 MeV步长 (12.0-0.8)/5600≈0.002
        n_E_d_bins=560,
        n_E_p_bins=560,)
junosyst.get_obs_spectrum(gseed)

# junosyst.initial_params_inverted对应逆序拟合，junosyst.initial_params_normal对应正序拟合
m = Minuit(junosyst.chi2, junosyst.NO_params_NO, name=junosyst.names)

m.tol = 0.01
# m.fixed["dmsq31"] = True
results = m.migrad()
display(results)

```

```python
from torch_helper import fit, ReactorLoss
from src.fitter.analysis.juno_syst import JunoSyst
import torch

# 1. 设置环境（与jingqin_SGD.ipynb一致）
torch.set_default_dtype(torch.float64)
year = 50/365.25  # 数据量设置

# 2. 创建JUNO系统分析对象
juno_syst = JunoSyst(year,        
    n_E_nu_bins=5600,  # 0.002 MeV步长
    n_E_dep_bins=5600, 
    n_E_d_bins=560,
    n_E_p_bins=560,
)

# 3. 生成观测数据
obs_spectrum = juno_syst.get_obs_spectrum(seed=42)

# 4. 使用SGD进行拟合
x, y, t, lr, reactor_loss, edm_values = fit(
    juno_syst,
    initial_params=None,  # 使用默认参数
    n_iter=1000,         # 迭代次数
    lr=1e-2,             # 保守的学习率
    use_edm=True,        # 使用EDM收敛判断
    edm_tolerance=1e-6,  # EDM阈值
    edm_method='approximate'  # 或 'exact', 'hybrid'
)

# 5. 分析结果
osc_params = reactor_loss.get_oscillation_params()
print("拟合的振荡参数:")
for name, value in osc_params.items():
    print(f"  {name} = {value:.6e}")

# 6. 查看拟合历史
print(f"最终损失: {y[-1]:.6f}")
print(f"总迭代数: {len(x)}")
```

```python
import torch
import numpy as np
import time
from functools import lru_cache

def is_hashable(var):
    try:
        hash(var)
        return True
    except TypeError:
        return False

ncalls = 0
@lru_cache(maxsize=5)
def test(var):
    global ncalls
    ncalls += 1
    print("in func ", var)
    return var

def run(var1, var2):
    print("if can cache: ", is_hashable(var1))
    try:
        print("check ", var1)
        test(var1)
        print("ncalls: ", ncalls)
    except:
        print('error')
    try:
        print("check ", var1)
        test(var1)
        print("ncalls: ", ncalls)
    except:
        print('error')
    try:
        print("check ", var2)
        test(var2)
        print("ncalls: ", ncalls)
    except:
        print('error')


# 测试1：torch.Tensor 行！
print("=======> test1: torch.Tensor")
ncalls = 0
list1 = [1, 2, 3]
var1 = torch.tensor(list1)
var2 = torch.tensor([4, 5, 6])
run(var1, var2)
var2[1] = 100
print("check ", var2)
test(var2)
print("ncalls: ", ncalls)
var3 = torch.tensor(list1)
test(var3)
print("ncalls: ", ncalls)
var4 = torch.tensor([4, 5, 6])
test(var4)
print("ncalls: ", ncalls)
```
